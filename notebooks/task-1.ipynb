#### **What is the overall Loss Ratio (TotalClaims / TotalPremium) for the portfolio? How does it vary by Province, VehicleType, and Gender?**

```python
# Calculate overall Loss Ratio for the portfolio
overall_loss_ratio = df['TotalClaims'].sum() / df['TotalPremium'].sum()
print(f"Overall Loss Ratio: {overall_loss_ratio:.4f}")

# Loss Ratio by Province
province_loss_ratio = df.groupby('Province').apply(lambda x: x['TotalClaims'].sum() / x['TotalPremium'].sum())
print("\nLoss Ratio by Province:")
print(province_loss_ratio)
## plot barplot 

# Loss Ratio by VehicleType
vehicle_type_loss_ratio = df.groupby('VehicleType').apply(lambda x: x['TotalClaims'].sum() / x['TotalPremium'].sum())
print("\nLoss Ratio by VehicleType:")
print(vehicle_type_loss_ratio)
## plot barplot 

# Loss Ratio by Gender
gender_loss_ratio = df.groupby('Gender').apply(lambda x: x['TotalClaims'].sum() / x['TotalPremium'].sum())
print("\nLoss Ratio by Gender:")
print(gender_loss_ratio)
## plot barplot 
```

```python
# Plot Loss Ratio by Province
plt.figure(figsize=(10, 5))
sns.barplot(x=province_loss_ratio.index, y=province_loss_ratio.values, palette="Blues_d")
plt.title("Loss Ratio by Province")
plt.ylabel("Loss Ratio")
plt.xlabel("Province")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot Loss Ratio by VehicleType
plt.figure(figsize=(8, 5))
sns.barplot(x=vehicle_type_loss_ratio.index, y=vehicle_type_loss_ratio.values, palette="Greens_d")
plt.title("Loss Ratio by Vehicle Type")
plt.ylabel("Loss Ratio")
plt.xlabel("Vehicle Type")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

# Plot Loss Ratio by Gender
plt.figure(figsize=(6, 4))
sns.barplot(x=gender_loss_ratio.index, y=gender_loss_ratio.values, palette="Purples_d")
plt.title("Loss Ratio by Gender")
plt.ylabel("Loss Ratio")
plt.xlabel("Gender")
plt.tight_layout()
plt.show()
```

```python
def plot_boxplots(df, numerical_features, figsize=(15, 8), whis=1.5, palette='viridis'):
    """
    Plots enhanced box plots for numerical features with outlier analysis.
    
    Parameters:
        df (pd.DataFrame): DataFrame containing the data
        numerical_features (list): List of numerical column names
        figsize (tuple): Figure size (width, height)
        whis (float): Whisker length (IQR multiplier)
        palette (str): Color palette name
    
    Returns:
        dict: Outlier statistics for each feature
    """
    import numpy as np
    
    # Calculate subplot grid dimensions
    n_cols = min(3, len(numerical_features))  # Ensure we don't create empty subplots
    n_rows = int(np.ceil(len(numerical_features) / n_cols))
    
    # Create figure with subplots
    fig, axes = plt.subplots(n_rows, n_cols, figsize=figsize)
    
    # If only one feature, axes won't be an array
    if len(numerical_features) == 1:
        axes = np.array([axes])
    
    # Flatten axes array for easy iteration
    axes = axes.flatten()
    
    # Get a color palette with enough colors
    colors = sns.color_palette(palette, n_colors=len(numerical_features))
    
    # Dictionary to store outlier info
    outliers_info = {}
    
    # Plot each numerical feature
    for i, (col, ax) in enumerate(zip(numerical_features, axes)):
        # Create boxplot with enhanced styling
        sns.boxplot(
            y=df[col],
            ax=ax,
            color=colors[i],
            width=0.4,
            whis=whis,
            flierprops={
                'marker': 'o',
                'markersize': 5,
                'markerfacecolor': 'red',
                'markeredgecolor': 'orange',
                'alpha': 0.5
            }
        )
        
        # Calculate and annotate outliers
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - whis * iqr
        upper_bound = q3 + whis * iqr
        
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
        outliers_pct = (len(outliers) / len(df)) * 100
        
        # Store outlier info
        outliers_info[col] = {
            'count': len(outliers),
            'percentage': outliers_pct,
            'lower_bound': lower_bound,
            'upper_bound': upper_bound
        }
        
        # Add annotations
        ax.set_title(f'{col}\nOutliers: {len(outliers)} ({outliers_pct:.1f}%)')
        ax.set_ylabel('')
        ax.grid(True, linestyle='--', alpha=0.7)
        
        # Rotate x-axis labels if needed
        plt.setp(ax.get_xticklabels(), rotation=45)
    
    # Hide unused subplots
    for j in range(len(numerical_features), len(axes)):
        axes[j].set_visible(False)
    
    plt.tight_layout()
    plt.show()
    
    return outliers_info

numerical_features = df.select_dtypes(include="float").columns.to_list()
outliers_infomation = plot_boxplots(df, numerical_features=numerical_features)
```
